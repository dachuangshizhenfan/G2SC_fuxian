```python
import json
import pandas as pd
pd.set_option('max_columns',1000)
pd.set_option('max_row',1000)
pd.set_option('max_colwidth',1000)

with open('abc.json','r',encoding='utf8')as fp:
    json_data = json.load(fp)
    print('这是文件中的json数据：',json_data)
    print('这是读取到文件数据的数据类型：', type(json_data))
    print(json_data.keys())
with open('abc.json', 'w') as file:
    json.dump(json_data, file, indent = 4)
# 读取json文件内容,返回字典格式

print(json_data['name'])
print(json_data['directed'])
print(json_data['strict'])
print(json_data['label'])
print(json_data['_subgraph_cnt'])
del json_data['objects'][0]
print(json_data['objects'])
print(json_data['edges'])
#abc['objects'][0]
#df = pd.read_json("abc.json",encoding="utf-8", orient='records')
#print(df)
df_nodes = pd.DataFrame(json_data['objects'])
print(df_nodes)
df_edges = pd.DataFrame(json_data['edges'])
print(df_edges)
```

```python
df_edges.drop(df_edges.index[(df_edges['style'] == 'bold')], inplace=True)
df=df_edges.assign(iscover=0,new_tail=0,new_head=0,new_gvid=0)
df_nodes=df_nodes.assign(color=-1,d=0,f=0,new_gvid=0)

b,c = df_nodes.iloc[df_nodes.index[df_nodes['fillcolor']=='aquamarine'].to_list()[0]], df_nodes.iloc[0]
#print(b)
#print(c)
#print(b['_gvid'])

queue=[]  #建立队列
queue.append(b)
data=[]  #记录已经遍历过的点
data.append(int(b['_gvid']))
clock = 0
while queue:
    n=queue.pop(0)  # 队列先进先出
    next_nodes_gvid = df_edges.loc[df_edges['tail']==n['_gvid']]
    #print(next_nodes_gvid)
    nodes = []
    for index, row in next_nodes_gvid.iterrows():
        a = row['head']
        nodes.append(df_nodes.iloc[df_nodes.index[df_nodes['_gvid']==a].to_list()[0]])
    #print(nodes)
    for j in nodes:
        #print(j)
        #print(data)
        if j['_gvid'] not in data:
            queue.append(j)
            data.append(int(j['_gvid']))
    ##print(n)
    ngvid = n['_gvid']
    #print(ngvid)
    index0 = df_nodes[df_nodes['_gvid']==ngvid].index.to_list()[0]
    #print(index0)
    df_nodes.at[index0,'new_gvid'] = clock
    clock = clock + 1

print(df_nodes)
    

#df_nodes=df_nodes.reindex(index=['2','0','1','3','4'])
#print(df)
#print(df_nodes)
#print(df.shape[0])


graph=df['tail'].value_counts()

df_nodes.sort_values(by="new_gvid" , inplace=True, ascending=True)
df_nodes = df_nodes.reset_index(drop=True)

print(df_nodes)

for index, row in df.iterrows():
    tmp_old_tail = row['tail']
    tmp_new_tail=df_nodes.loc[df_nodes['_gvid']==tmp_old_tail]['new_gvid']
    df.at[index,'new_tail'] = tmp_new_tail
    
    tmp_old_head = row['head']
    tmp_new_head=df_nodes.loc[df_nodes['_gvid']==tmp_old_head]['new_gvid']
    df.at[index,'new_head'] = tmp_new_head
    
df.sort_values(by="new_head" , inplace=True, ascending=True)
df = df.reset_index(drop=True)

new_edge_clock = 0
for index, row in df.iterrows():
    df.at[index,'new_gvid'] = new_edge_clock
    new_edge_clock = new_edge_clock + 1

print(df)
```

```python
#start_node = int(df_nodes.loc[df_nodes['shape']=='diamond']['_gvid'])


iscover_sum=0
t=0
graph_to_sequence = []


def dfs_visit(index):


    df_nodes.at[index,'color'] = 0
    global t
    t = t + 1
    df_nodes.at[index,'d'] = t
    #print(df_nodes)
    a = df_nodes.iloc[index]['new_gvid']
    subgraph = df.loc[df['new_tail']==a]
    #print(subgraph)
    for index2, row2 in subgraph.iterrows():
        tmp_graph_to_sequence = []
        if(row2['iscover']==1):
            break
        df.at[index2,'iscover'] =1
        b=row2['new_head']
        #print(b)
        val = int(df_nodes.loc[df_nodes['new_gvid']==b]['color'])
        #print(val)
        if val == 0:
            print(str(row2['new_tail'])+" "+str(row2['new_head'])+"后向边")
            tmp_graph_to_sequence.clear()
            tmp_graph_to_sequence.append(row2['new_tail'])
            tmp_graph_to_sequence.append(row2['new_head'])
            tmp_graph_to_sequence.append(row2['new_gvid'])
            tmp_graph_to_sequence.append(row2['label'])
            tmp_graph_to_sequence.append(row2['style'])
            graph_to_sequence.append(tmp_graph_to_sequence)
            #print(tmp_graph_to_sequence)
            #print(graph_to_sequence)
        else:
            if row2['style']=="dotted":
                print(str(row2['new_tail'])+" "+str(row2['new_head'])+"控制依赖")
                tmp_graph_to_sequence.clear()
                tmp_graph_to_sequence.append(row2['new_tail'])
                tmp_graph_to_sequence.append(row2['new_head'])
                tmp_graph_to_sequence.append(row2['new_gvid'])
                tmp_graph_to_sequence.append(row2['label'])
                tmp_graph_to_sequence.append(row2['style'])
                graph_to_sequence.append(tmp_graph_to_sequence)
                #print(tmp_graph_to_sequence)
                #print(graph_to_sequence)
                dfs_visit(df_nodes.loc[df_nodes['new_gvid']==b].index.to_list()[0])
            elif row2['style']=="solid":
                print(str(row2['new_tail'])+" "+str(row2['new_head'])+"数据依赖")
                tmp_graph_to_sequence.clear()
                tmp_graph_to_sequence.append(row2['new_tail'])
                tmp_graph_to_sequence.append(row2['new_head'])
                tmp_graph_to_sequence.append(row2['new_gvid'])
                tmp_graph_to_sequence.append(row2['label'])
                tmp_graph_to_sequence.append(row2['style'])
                graph_to_sequence.append(tmp_graph_to_sequence)
                #print(tmp_graph_to_sequence)
                #print(graph_to_sequence)
                dfs_visit(df_nodes.loc[df_nodes['new_gvid']==b].index.to_list()[0])
    t = t + 1
    df_nodes.at[index,'f']=t
    df_nodes.at[index,'color']=1


for index, row in df_nodes.iterrows():
    #print(index)
    if df_nodes.iloc[index]['color']==-1:
        #print(index)
        #print(row)
        #print("aaaaaaaaaaaaaaaaaaaaaaaaaaa")
        dfs_visit(index)
```
